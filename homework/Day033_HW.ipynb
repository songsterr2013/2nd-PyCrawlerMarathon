```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get('https://github.com/new')
soup=BeautifulSoup(r.text,'lxml')
print(soup.text)
```

```python
#關鍵在於觀察登入前後的cookies變化，多了些什麼，然後把cookies用進去就ok了
cookies={
    'logged_in':'yes',
    'user_session':'Z1qHrpTOJ7uGUmJgxpHXkNkZkank5MmKDh5C2'
}#為安全考慮，以上user_session經修改
r = requests.get('https://github.com/new', cookies=cookies)
soup = BeautifulSoup(r.text, 'lxml')
print(soup.text) # 加上檢查的 cookies ，就可以正確取得資料
```
