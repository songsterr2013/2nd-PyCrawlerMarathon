### 指定你想要觀看的放映地區，查詢有上映電影的場次日期

```python
# 指定放映地區
area_id = 28
```

```python
# 向網站發送請求
url = 'https://movies.yahoo.com.tw/movietime_result.html'
payload = {'movie_id':str(movie_id), 'area_id':str(area_id)} #指定哪一部+指定放映地區
resp = requests.get(url, params=payload)
resp.encoding = 'utf-8'

soup = BeautifulSoup(resp.text, 'lxml')
movie_date = soup.find_all("label", attrs={'for':re.compile("date_[\d]")})

# 列印播放日期
for date in movie_date:
    print("%s %s" % (date.p.string, date.h3.string))
```

### 最後指定觀看的日期，查詢並列印出放映的電影院、放映類型(數位、3D、IMAX 3D...)、放映時間等資訊。

```python
# 選定要觀看的日期
date = "2020-03-12"
```

```python
# 向網站發送請求，獲取上映的電影院及時間資訊
url = "https://movies.yahoo.com.tw/ajax/pc/get_schedule_by_movie"
payload = {'movie_id':str(movie_id),#哪一部
           'date':date,#哪一天  2019
           'area_id':str(area_id),#哪個地區 台北
           'theater_id':'',
           'datetime':'',
           'movie_type_id':''}

resp = requests.get(url, params=payload)
#print(resp.json()['view'])  # 若有需要，列印出json原始碼

soup = BeautifulSoup(resp.json()['view'], 'lxml')
html = soup.find_all("ul", attrs={'data-theater_name':re.compile(".*")})
print(html)
```

```python
#  試著從上一步驟回傳的電影院資料中，擷取電影院名稱、影片放映類型以及時間表
for i in html:
    
    #電影院名稱data-theater_name="台北美麗華大直影城"
    print(i['data-theater_name'])#他是tag裡的東西
    
    #影片放映類型
    print(i.find('span',class_="tapR").text)#他不是tag裡面的東西，所以要find他的text
    
    #時間表
    for d in i.find_all('input',class_="gabtn"):#他有很多個，所以要先全找出來 find_all再一個一個取
        print(d['value'])#他是tag裡的東西
    
    print('======================================')

# 記得如果要找的是tag裡面的東西直接用['']就好
# 如果是tag'包住'的東西，那就是要find.('',='')
```

# 下面是一些個人的練習

```python
#個人備註:
#點選「上映中」的分頁，先確認上映中的電影總共有幾筆，然後逐頁爬取。
#=========================================
import requests
from bs4 import BeautifulSoup

url='https://movies.yahoo.com.tw/movie_intheaters.html'
soup=BeautifulSoup(requests.get(url).text,'lxml')
print(soup.prettify())
```

```python
print('上映中的電影:',soup.find('div',class_="release_time _c").find('p').text)
```

```python
#個人備註
#利用while迴圈做法去換頁
#=============================================
count=1  #設定他換頁
page=1   #設定他換頁顯示
num=1
while True:
    
    #下面這個soup的page={}會隨著他的count而一直加上去，達到不停換頁的效果
    soup=BeautifulSoup(requests.get('https://movies.yahoo.com.tw/movie_intheaters.html?page={}'.format(count),'lxml').text)
    
    #每一頁中都要print以下的這些資料(名稱、期待度、滿意度)，加上try,except是因為有些電影或許還沒有滿意度/期待度
    for i in soup.find('ul',class_="release_list").find_all('li'):      

        print('{}'.format(num),i.find('div',class_='release_movie_name').find('a').text.replace(' ',''))

        try:
            print(i.find('dt').find('div',class_='level_name').text,i.find('div',class_='leveltext').text.replace(' ','').replace('\n','').replace('\t',''))
        except:
            print('NaN')
        try:
            print(i.find('dd').find('div',class_='level_name').text,i.find('div',class_='leveltext').text.replace(' ','').replace('\n','').replace('\t',''))
        except:
            print('NaN')
        num+=1 #注意他要在這個for裡面就要+1
        
    count+=1 #在while裡而+1，因為要換頁了
    
    print('========================第{}頁結束=========================='.format(page))  #換頁分隔線
    
    page+=1 
    
    if count ==6:  #設定條件，因為知道他只有5頁，所以設定到6就break，不然他的while true會換loop
        break
    else:
        continue
```
