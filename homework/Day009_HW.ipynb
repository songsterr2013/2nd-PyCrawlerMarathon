# 作業

以正確的副檔名下載網頁中的圖片
範例網頁:https://www.ptt.cc/bbs/Beauty/M.1556291059.A.75A.html

Note：因為 PTT 會詢問「是否滿 18 歲」，這邊可以用 cookies 繞過

`requests.get(URL, cookies={'over18': '1'}`


```python
import requests
import os #為了等下創新FILE

from bs4 import BeautifulSoup
from PIL import Image #處理圖片的一個套件
```

```python
#個人備註:
#cookies={'over18': '1'}
#===================================
url = 'https://www.ptt.cc/bbs/Beauty/M.1556291059.A.75A.html'
resp = requests.get(url, cookies={'over18': '1'})
soup = BeautifulSoup(resp.text)
print(soup.prettify())
```

```python
# 決定要儲存的資料夾
output_dir = 'downloads'

# 假如資料夾不存在就新增一個資料夾，跟之前教的方法不一樣，可學
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

    
```

```python
#檢查
mypath = ('.')
files =  os.listdir(mypath)


if 'downloads' in files:
    print('[O]')
else:
    print('[X]')
```

# 以下為範例做法，後面會加上個人理解的過程

```python
image_tags = soup.find(id='main-content').findChildren('a', recursive=False)
for img_tag in image_tags:

    if 'imgur' not in img_tag['href']:
        continue
        
    img_id = img_tag['href'].split('/')[-1]
    
    img_url = 'https://i.imgur.com/{}.jpg'.format(img_id)

    with requests.get(img_url, stream=True) as r:
        r.raise_for_status()

        img = Image.open(r.raw)
        img_savename = '{outdir}/{img_id}.{img_ext}'.format(
            outdir=output_dir, img_id=img_id, img_ext=img.format.lower())
        img.save(img_savename)
        print('Save image {}'.format(img_savename))
```

# 莫凡教學，用正則去找也可以

```python
import re
```

```python
#莫凡教學

#正則表達式
#  '.'=任何字符，除\n
#  '*'=重复零次或多次


#compile 
#可重複地對使用這個正則，先将正则 compile 进一个变量, 
#比如 compiled_re, 然后直接使用这个 compiled_re 来搜索.



                     #就這句變了而已，但這句跟原來找到的東西會一模一樣
sss=soup.find_all('a',{'href':re.compile('https://imgur.com/.*')})





for img_tag in sss:

    if 'imgur' not in img_tag['href']:
        continue
        
    img_id = img_tag['href'].split('/')[-1]
    
    img_url = 'https://i.imgur.com/{}.jpg'.format(img_id)

    with requests.get(img_url, stream=True) as r:
        r.raise_for_status()

        img = Image.open(r.raw)
        img_savename = '{outdir}/{img_id}.{img_ext}'.format(
            outdir=output_dir, img_id=img_id, img_ext=img.format.lower())
        img.save(img_savename)
        print('Save image {}'.format(img_savename))
```

# 什麼是recursive?

```python
# 個人備註

#https://medium.com/pyladies-taiwan/%E5%BE%9Edcard%E7%B6%B2%E7%AB%99%E7%9C%8B%E7%88%AC%E8%9F%B2%E5%85%A5%E9%96%80-65105f0ddac

# recursive是否「遞迴處理」子屬性(True or False)
#Beautiful Soup會查詢目前tag的子孫節點，如果只想要搜詢tag的直接子節點,參數可以使用recursive=False .
#下列為例子：
#<html> 
#<head> 
#  <title> 
#   FeiFei的主頁
#  </ title> 
#</ head> 
#...




#是否使用 recursive 參數的搜索结果:



#soup.html.find_all（“title”）
#         [<title> FeiFei的主頁</ title>]

#soup.html.find_all（“title”，recursive = False）
#        []


#因為<title>標籤在 <html> 標籤下, 
#但並不是直接的子節點， <head>標籤才是直接子節點， 
#在允許查询所有後代節點時 Beautiful Soup 能夠找到 <title> 標籤， 
#但是使用了 recursive=False參數之後，只能查找直接子標籤，
#這樣就查不到 <title> 標籤。
#=========================================================
```

```python
# 個人備註:
#這是範例的做法，對我而言有點太進階，但仍然把它拆開，每個都去了解他
```

```python
#個人備註

#尋找第一個「id='main-content'」
#然後從裡面尋找每一個(findChildren=find_all)「'a'」
#recursive=False就是說「只能查找直接子標籤」/「不是子節點的東西不要」

#   ****重點****
#Beautiful Soup會查詢目前tag的子孫節點，
#如果只想要搜詢tag的直接子節點,參數可以使用recursive=False 

#因為<title>標籤在 <html> 標籤下, 
#但並不是直接的子節點， <head>標籤才是直接子節點， 
#在允許查询所有後代節點時 Beautiful Soup 能夠找到 <title> 標籤， 
#但是使用了 recursive=False參數之後，只能查找直接子標籤，
#這樣就查不到 <title> 標籤。
#====================================================

#定位所有圖片的 tag
image_tags = soup.find(id='main-content').findChildren('a', recursive=False)
image_tags
```

```python
#個人備註
#這邊抽取每一個herf後的網址
#有這一步是因為想要看看沒有.split('/')[-1]會如何
#==========================================================
for i in image_tags:
    
   # 取得所有圖片在第三方服務的 id
    if 'imgur' not in i['href']:
        continue
        
    img_id = i['href']
    print(img_id)
```

```python
#個人備註
#這邊的.split('/')[-1]
#原來是代表「分開」，「/」為關鍵字，「-1」的意思即取最後的一段
#如果他是   「.split('/')」   的話:
#['https:', '', 'imgur.com', 'Cgb5oo1']代表每看到「/」就分隔開來
#所以  「.split('/')[-1]」   就是取最後一塊
#取完之後做接下來format的動作
#==========================================================
for img_tag in image_tags:
    
   # 取得所有圖片在第三方服務的 id
    if 'imgur' not in img_tag['href']:
        continue
        
    img_id = img_tag['href'].split('/')[-1]
    print(img_id)
```
