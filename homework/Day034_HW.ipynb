## 作業目標

* 目前程式中的 proxy_ips 是手動輸入的，請根據 https://free-proxy-list.net/ 寫一個可自動化抓取可用 Proxy 的 proxy_ips。




```python
import requests
from bs4 import BeautifulSoup
import random
```

```python
r=requests.get('https://free-proxy-list.net/')
soup=BeautifulSoup(r.text,'lxml')
print(soup.prettify())
```

```python
proxy_ips = []
for i in soup.find('tbody').find_all('tr'):
    for tds in i.find('td'):
        proxy_ips.append(tds)
```

```python
proxy_ips
```

```python
#感覺這個try,except就是用來試說，有可能有些ip被搶先用了
for i in range(10):
    ip = random.choice(proxy_ips)
    print('Use', ip)
    try:
        resp = requests.get('http://ip.filefab.com/index.php',
                        proxies={'http': ip, 'https': ip}, timeout=2)
        soup = BeautifulSoup(resp.text, 'lxml')
        print(soup.find('h1', id='ipd').text.strip())
    except:
        print('Fail')
```

```python
#拿剛剛弄回來的IP試試看再檢查一次他的IP
trytrylook = requests.get('http://ip.filefab.com/index.php',
                proxies={'http': '63.232.67.5'})
soup2 = BeautifulSoup(trytrylook.text, 'lxml')

soup2.find('h1', id='ipd').text.strip()
```
