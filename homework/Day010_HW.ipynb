# =====================

#  requests做法

```python
#個人備註:
#用知乎去獲取他的網頁資訊
#因為之前知乎用requests的方法他要求user-agent
#現在不加是沒有成功的
#========================================

import requests as rq

r=rq.get('https://www.zhihu.com/topic/19552832/hot')
print(r.text)
```

```python
#個人備註:
#加入Headers後成功
#========================================

import requests as rq
from bs4 import BeautifulSoup as btfs 
headers= {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'}
r=rq.get('https://www.zhihu.com/topic/19552832/hot',headers= headers)
print(r.text)
```

```python
print('request後取出的型態為:',type(r.text))
```

# BeautifulSoup做法

```python
#個人備註:
#轉為btfs可剖析的格式
#========================================


soup = btfs(r.text,'lxml')
print(soup.prettify())
```

```python
print('變成btfs後的型態為:',type(soup))
```

# Grab + PyQuery

# grab做法

```python
#個人備註:
#一樣用知乎去獲取他的網頁資訊
#用grab的方法好像直接就讓我們抓取了
#========================================


from grab import Grab
g = Grab()
resp = g.go('https://www.zhihu.com/topic/19552832/hot')
print(resp.body)
```

```python
#個人備註:

#去查了一下grab的介紹，少之有之，看他的官網介如好像有提到
#HTTP and SOCKS proxy with/without authorization
#不知道是否因為這樣所以不用headers就可以直接通過
#但問題是他不是str? 那要怎麼轉化他裡面的東西成為可供閱讀的中文?

#型態不一樣，這是bytes，感覺在Grab在處理資料時會跟之前所用的request跟btfs不一樣
#========================================
print('透過grab取出的型態為:',type(resp) , type(resp.body))
```

#  PyQuery做法

```python
#個人備註:
#看教學:   python爬蟲神器PyQuery的使用方法
#https://codertw.com/%E7%A8%8B%E5%BC%8F%E8%AA%9E%E8%A8%80/119571/

#指定檢索第一個title內的文字
#且自動轉為中文?
#========================================

from pyquery import PyQuery as pq
doc = pq(resp.body)
title = doc('title')
print('變成了{}套件:'.format(type(title)), '\ntitle中的文字為:',title.text())
```
