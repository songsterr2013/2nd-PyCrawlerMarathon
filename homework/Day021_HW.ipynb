## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser=webdriver.Chrome()
browser.get('https://www.ettoday.net/news/news-list.htm')
```

```python
import datetime
today = datetime.date.today()
exactly_toady=(today.strftime("%Y/%m/%d"))
print(exactly_toady,type(exactly_toady))
```

```python
import re
#這是按目前我所能夠做的去做
soup=BeautifulSoup(browser.page_source, "lxml")
for i in soup.find('div',class_='part_list_2').find_all('h3'):
    if re.compile('(\d\d):(\d\d)').sub('',i.find('span',class_="date").text.replace(' ',''))!=exactly_toady:
        break
    print(i.text)
```

```python
#以下是原本應該要達到的:

#這邊有個問題是，這個練習基於「他的今日新聞量達到可以往下捲動的程度」,
#但問題1是他最多只能顯示780個新聞，假如我在某天的晚上看，他的當日新聞可能就會超過780,
#可能無法看到早上或半夜的新聞

#問題2是如果我在接近00:00後看，他的「當日新聞量」太少，
#根本不用捲動，我就根本不需要使用selenium

#這是假設他今天不超過780則，且又可以往下捲動的情況:

import time
import re
while True:
    time.sleep(0.5)
    soup=BeautifulSoup(browser.page_source, "lxml")
    date=re.compile('(\d\d):(\d\d)').sub('',soup.find('div',class_='part_list_2').find_all('h3')[-1].find('span',class_="date").text.replace(' ',''))
    print(date)
    browser.execute_script("window.scrollTo(0, 9999999999999);")
    if date!=exactly_toady:
        print('-------------------------')
        break    
#因為他不用捲動他最後一則就是3月18了，但其實3月19已經有了
        
        
#然後再用for的方法把超過的文章過濾掉，
#因為他的find_all('h3')[-1]其實是在做一個確認需不需要繼續往下捲動的動作而已
```

## 2. 取出現在時間兩小時內的新聞

```python
#1. timedelta 對象
#timedelta 對象表示二個日期或時間之間的時間差

#class datetime.timedelta
#(days=0, seconds=0, microseconds=0, milliseconds=0, 
#minutes=0, hours=0, weeks=0)

#原文網址：https://kknews.cc/code/3zokk5g.html



#2.Python時間strptime（）函數根據指定的格式把一個時間字符串解析為時間元組。
#http://tw.gitbook.net/python/time_strptime.html
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser=webdriver.Chrome()
browser.get('https://www.ettoday.net/news/news-list.htm')
```

```python
from datetime import datetime, timedelta
two_hours_ago=datetime.now()-timedelta(hours=2)
print('兩小時前的時間為:',datetime.now()-timedelta(hours=2))
```

```python
from datetime import datetime, timedelta
#取得網頁資訊:
soup2=BeautifulSoup(browser.page_source, "lxml")

table_two_hours_ago=[]
#篩選
for i in soup2.find('div',class_='part_list_2').find_all('h3'):
  #愈貼近現在的時間愈大，所以如果貼文的時間 「大於」 2小時前的時間就print出貼文
  #也就是說如果是比他小的，就等於他不是「兩小時內的文章」
    if datetime.strptime(i.find('span',class_="date").text,'%Y/%m/%d %H:%M')>two_hours_ago:
        table_two_hours_ago.append([i.find('span',class_="date").text,i.find('a',target="_blank").text])
```

```python
import pandas as pd
columns=['日期','兩小時內的文章標題']
data=pd.DataFrame(table_two_hours_ago,columns=columns)
data
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from datetime import datetime, timedelta
#3日前的下午 3點 與 5點:

#strftime由日期格式轉化為字串格式    type** str
#strptime由字串格式轉化為日期格式    type** datetime.datetime
#所以是先轉成字串，加上15，再一起轉化為日期
three_days_ago= datetime.now() - timedelta(3)
        #用strftime，三天前的時間重新排列為只顯示「年月日」，現在加上15，
                                   #然後再用strptime，指定他為'年月日時去排列
threepm=datetime.strptime(three_days_ago.strftime("%Y/%m/%d") + ' 15', '%Y/%m/%d %H')
fivepm=datetime.strptime(three_days_ago.strftime("%Y/%m/%d") + ' 17', '%Y/%m/%d %H')
print('三天前下午三點:',threepm)
print('三天前下午五點:',fivepm)
print(type(threepm),type(fivepm))
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser=webdriver.Chrome()
browser.get('https://www.ettoday.net/news/news-list.htm')
```

```python
import time

while True:   
    #先來個大包圍
    for i in range(50):#知道他大概捲動個50次就會到底，反正捲到底一定不會錯
        browser.execute_script("window.scrollTo(0, 9999999999999);")
        time.sleep(0.5)
    #把大包圍後的結果存成bs4格式
    soup3=BeautifulSoup(browser.page_source,"lxml")
    #用時間去抽取指定的新聞
    #記得這邊「時間變得可以計算是因為把他格式轉成datetime.datetime」
    #原來的str不具有任何可以計算的性質，加上datetime.strptime()就OK了
    for i in soup3.find('div',class_='part_list_2').find_all('h3'):
        news=datetime.strptime(i.find('span',class_="date").text,'%Y/%m/%d %H:%M')
        if threepm<news<fivepm:
            print(i.find('span',class_="date").text,i.find('a',target="_blank").text)        
    
    break
```
