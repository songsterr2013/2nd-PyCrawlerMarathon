```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

# å€‹äººå‚™è¨»
#é€™é‚Šæœ‰å€‹%sæ–¹æ³•ä»¥å¾Œå¯ä»¥è©¦è©¦çœ‹
#==================================================

import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
js=json.loads(r.text)

count=0
for i in js:
    print(i['id'])
    count+=1
print('ç¸½å…±å›å‚³%sè³‡æ–™'%count)




#ç­”:ä¸€æ¬¡å…±å›å‚³30ç­†è³‡æ–™ï¼Œ
```

```python
# å€‹äººå‚™è¨»
#å› ç‚ºjsonæ˜¯å­—å…¸å½¢å¼(key value)ï¼Œæ‰€ä»¥ç”¨ key valueçš„å½¢å¼å»æ‰¾å‡ºæœ‰å¹¾å€‹keys
#==================================================

print('æ¯ä¸€ç­†è³‡æ–™åŒ…å«ä»¥ä¸‹æ¬„ä½:')
num=js[1].keys()
count_1 = 0
for i in num:
    print(i)
    count_1+=1

print('å…±æœ‰%så€‹æ¬„ä½'%count_1)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

# å€‹äººå‚™è¨»
#title
#createdAt
#commentCount
#likeCount
#==================================================
count = 1
for i in js:
    print(count,'æ¨™é¡Œ: ',i['title'])
    count+=1
    print('è²¼æ–‡æ™‚é–“: ',i['createdAt'])
    print('ç•™è¨€äººæ•¸: ',i['commentCount'])
    print('æŒ‰è®šäººæ•¸: ',i['likeCount'])
    print('================')
```

```python
# å€‹äººå‚™è¨»
#PANDASæ–¹æ³•
#=================================================

import pandas as pd

data = pd.DataFrame(js)

#é¸å–æˆ‘å€‘éœ€è¦ä»€éº¼æ±è¥¿
allineed = data[['title','createdAt','commentCount','likeCount']]
#å°‡æˆ‘å€‘éœ€è¦çš„æ±è¥¿æ”¹å€‹åå­—
allineed.columns = ['æ¨™é¡Œ:','è²¼æ–‡æ™‚é–“:','ç•™è¨€äººæ•¸:','æŒ‰è®šäººæ•¸:']
#å‘ˆç¾å‡ºä¾†
allineed
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
# å€‹äººå‚™è¨»
#https://www.dcard.tw/_api/forums/pet/posts?popular=true  é€™æ˜¯ç†±é–€
#https://www.dcard.tw/_api/forums/pet/posts?popular=false é€™æ˜¯éç†±é–€å³æœ€æ–°
#ç™¼ç¾ä»–APIçš„ç¶²å€çš„é‚è¼¯å¾Œå°±å¯ä»¥ç²å¾—ä¸åŒå…§å®¹äº†
#å¯çœ‹é€™ç¯‡https://levirve.github.io/blog/2016/Dccard-crawler/

#é‡è¦
#å¦‚æœåœ¨æ”¹columnsçš„æ™‚å€™ç›´æ¥ç”¨:
#latestdata.columns=['æ¨™é¡Œ:','è²¼æ–‡æ™‚é–“:','ç•™è¨€äººæ•¸:','æŒ‰è®šäººæ•¸:']æœƒæœ‰bugèªª:
#Pandas doesn't allow columns to be created via a new attribute name
#ä½†æ˜¯å¦‚æœå»ºç«‹ä¸€å€‹æ–°è®Šæ•¸rename2 = ['æ¨™é¡Œ:','è²¼æ–‡æ™‚é–“:','ç•™è¨€äººæ•¸:','æŒ‰è®šäººæ•¸:']
#å†latestdata.columns=rename2å¥½åƒå°±æ²’äº‹äº†
#https://stackoverflow.com/questions/52129876/userwarning-pandas-doesnt-allow-columns-to-be-created-via-a-new-attribute-name
#==================================================


latest = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
latestjson = json.loads(latest.text)

latestdata = pd.DataFrame(latestjson, columns=['title','createdAt','commentCount','likeCount'])

rename2 = ['æ¨™é¡Œ:','è²¼æ–‡æ™‚é–“:','ç•™è¨€äººæ•¸:','æŒ‰è®šäººæ•¸:']

latestdata.columns=rename2
latestdata
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

# å€‹äººå‚™è¨»

#æ¥ä¸‹ä¾†ç†±é–€/éç†±é–€éƒ½å·²ç¶“ç”¨å¥½äº†
#.mean()çš„æ–¹æ³•(é€™æ–¹æ³•ä¸çŸ¥é“æ˜¯ä¸æ˜¯ã€Œå·²è™•ç†éçš„æ•¸æ“šã€çš„å¹³å‡å€¼ç®—æ³•)
#statistics.mean(data)
#https://docs.python.org/zh-tw/3/library/statistics.html#statistics.mean

#é‚„æœ‰roundçš„æ–¹æ³•å–å››æ¨äº”å…¥: round(x,y)
#==================================================
print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:',round(allineed[('ç•™è¨€äººæ•¸:')].mean(),2))
print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€:',round(allineed[('æŒ‰è®šäººæ•¸:')].mean(),2))
print('éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:',round(latestdata[('ç•™è¨€äººæ•¸:')].mean(),2))
print('éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€:',round(latestdata[('æŒ‰è®šäººæ•¸:')].mean(),2))
```

```python
# å€‹äººå‚™è¨»
#çœ‹åˆ°åˆ¥äººå¯ä»¥é€™æ¨£ç”¨ï¼Œä¹Ÿè©¦è©¦çœ‹
#'ä¸€æ®µæ–‡å­—:{}'.format(æƒ³è¦é‹ç®—çš„æ±è¥¿ï¼Œçµæœä¸Ÿåˆ°å‰é¢çš„{}è£¡é¢)
#==================================================
print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:{}'.format(round(allineed[('ç•™è¨€äººæ•¸:')].mean(),2)))
print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€:{}'.format(round(allineed[('æŒ‰è®šäººæ•¸:')].mean(),2)))
print('éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:{}'.format(round(latestdata[('ç•™è¨€äººæ•¸:')].mean(),2)))
print('éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€:{}'.format(round(latestdata[('æŒ‰è®šäººæ•¸:')].mean(),2)))
```

```python
# å€‹äººå‚™è¨»
#ç”¨numpyè©¦è©¦çœ‹ï¼Œå¦ä¸€ç¨®å–å¹³å‡æ•¸çš„æ–¹å¼ 
#np.mean([])
#==================================================
import numpy as np
print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:',round(np.mean(allineed[('ç•™è¨€äººæ•¸:')]),2))
print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:',round(np.mean(allineed[('æŒ‰è®šäººæ•¸:')]),2))
print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:',round(np.mean(latestdata[('ç•™è¨€äººæ•¸:')]),2)) 
print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:',round(np.mean(latestdata[('æŒ‰è®šäººæ•¸:')]),2)) 
```
