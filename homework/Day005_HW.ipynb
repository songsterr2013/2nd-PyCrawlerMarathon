1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

# 個人備註
#這邊有個%s方法以後可以試試看
#==================================================

import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
js=json.loads(r.text)

count=0
for i in js:
    print(i['id'])
    count+=1
print('總共回傳%s資料'%count)




#答:一次共回傳30筆資料，
```

```python
# 個人備註
#因為json是字典形式(key value)，所以用 key value的形式去找出有幾個keys
#==================================================

print('每一筆資料包含以下欄位:')
num=js[1].keys()
count_1 = 0
for i in num:
    print(i)
    count_1+=1

print('共有%s個欄位'%count_1)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

# 個人備註
#title
#createdAt
#commentCount
#likeCount
#==================================================
count = 1
for i in js:
    print(count,'標題: ',i['title'])
    count+=1
    print('貼文時間: ',i['createdAt'])
    print('留言人數: ',i['commentCount'])
    print('按讚人數: ',i['likeCount'])
    print('================')
```

```python
# 個人備註
#PANDAS方法
#=================================================

import pandas as pd

data = pd.DataFrame(js)

#選取我們需要什麼東西
allineed = data[['title','createdAt','commentCount','likeCount']]
#將我們需要的東西改個名字
allineed.columns = ['標題:','貼文時間:','留言人數:','按讚人數:']
#呈現出來
allineed
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
# 個人備註
#https://www.dcard.tw/_api/forums/pet/posts?popular=true  這是熱門
#https://www.dcard.tw/_api/forums/pet/posts?popular=false 這是非熱門即最新
#發現他API的網址的邏輯後就可以獲得不同內容了
#可看這篇https://levirve.github.io/blog/2016/Dccard-crawler/

#重要
#如果在改columns的時候直接用:
#latestdata.columns=['標題:','貼文時間:','留言人數:','按讚人數:']會有bug說:
#Pandas doesn't allow columns to be created via a new attribute name
#但是如果建立一個新變數rename2 = ['標題:','貼文時間:','留言人數:','按讚人數:']
#再latestdata.columns=rename2好像就沒事了
#https://stackoverflow.com/questions/52129876/userwarning-pandas-doesnt-allow-columns-to-be-created-via-a-new-attribute-name
#==================================================


latest = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
latestjson = json.loads(latest.text)

latestdata = pd.DataFrame(latestjson, columns=['title','createdAt','commentCount','likeCount'])

rename2 = ['標題:','貼文時間:','留言人數:','按讚人數:']

latestdata.columns=rename2
latestdata
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

# 個人備註

#接下來熱門/非熱門都已經用好了
#.mean()的方法(這方法不知道是不是「已處理過的數據」的平均值算法)
#statistics.mean(data)
#https://docs.python.org/zh-tw/3/library/statistics.html#statistics.mean

#還有round的方法取四捨五入: round(x,y)
#==================================================
print('熱門文章的「平均留言人數」:',round(allineed[('留言人數:')].mean(),2))
print('熱門文章的「平均按讚人數」:',round(allineed[('按讚人數:')].mean(),2))
print('非熱門文章的「平均留言人數」:',round(latestdata[('留言人數:')].mean(),2))
print('非熱門文章的「平均按讚人數」:',round(latestdata[('按讚人數:')].mean(),2))
```

```python
# 個人備註
#看到別人可以這樣用，也試試看
#'一段文字:{}'.format(想要運算的東西，結果丟到前面的{}裡面)
#==================================================
print('熱門文章的「平均留言人數」:{}'.format(round(allineed[('留言人數:')].mean(),2)))
print('熱門文章的「平均按讚人數」:{}'.format(round(allineed[('按讚人數:')].mean(),2)))
print('非熱門文章的「平均留言人數」:{}'.format(round(latestdata[('留言人數:')].mean(),2)))
print('非熱門文章的「平均按讚人數」:{}'.format(round(latestdata[('按讚人數:')].mean(),2)))
```

```python
# 個人備註
#用numpy試試看，另一種取平均數的方式 
#np.mean([])
#==================================================
import numpy as np
print('熱門文章的「平均留言人數」:',round(np.mean(allineed[('留言人數:')]),2))
print('熱門文章的「平均留言人數」:',round(np.mean(allineed[('按讚人數:')]),2))
print('熱門文章的「平均留言人數」:',round(np.mean(latestdata[('留言人數:')]),2)) 
print('熱門文章的「平均留言人數」:',round(np.mean(latestdata[('按讚人數:')]),2)) 
```
