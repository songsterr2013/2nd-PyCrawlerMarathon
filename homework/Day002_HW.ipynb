### 比較一下範例檔案中的「File I/O」與「CSV Reader」讀出來的內容有什麼差異

# 用「File I/O」讀出來

```python
# 個人備註:
#參考DAY1用OS去下載CSV檔
#另外也可參考SAMPLE做法
#=============================================================

from urllib.request import urlretrieve
import os

try:
    res = 'https://pycrawler-fileentity.cupoy.com/2nd/homework/data/1581304662075/example.csv?t=1581304958202'
    urlretrieve(res,'./example.csv')#urlretrieve下載東西用
    print('寫入成功！')
except:
    print('發生錯誤！')
```

```python
# 個人備註:
#檢查是否在指定位置
# . 指的是與當前檔案同一目錄中
#=============================================================

mypath = ('.')
files =  os.listdir(mypath)


if 'example.csv' in files:
    print('[O] ')
else:
    print('[X] ')
```

```python
# 個人備註:
#用FOR方法去檢查
#=============================================================
for i in files:
    print(i)
```

```python
# 個人備註:
#因為裡面有中文，所以後面要加encoding ="utf-8"
#如果不加:'cp950' codec can't decode byte 0xba in position 4: illegal multibyte sequence
#=============================================================

with open('./example.csv','r',encoding="utf-8") as cs:
    print(cs.read())
```

# 用「CSV Reader」讀出來

```python
# 個人備註:
#這邊如果直接在csv的函式裡面用encoding=utf-8好像不允許
#故先用WITH語法先把他設成一個變數

#為什麼要加newline=''???
#這裡在開啟 csv 檔案時加上了 newline='' 參數，這是為了讓資料中包含的換行字元可以正確被解析，
#所以建議在讀取 csv 檔案時都固定加入這個參數。

#=============================================================

import csv

with open ('./example.csv',newline="",encoding='utf-8') as file:
    rows= csv.reader(file)#不然其實在這個()裡就可以直接設定了
    for i in rows:
        print(i)
```

#### 答:使用「CSV Reader」相比起「File I/O」的方式讀出來的內容，具有逗號分隔值，將不同的欄位做為分隔

### 根據範例檔案的結果：

1. 取出班次一的每一個時間
2. 將班次一的每一個時間用一種資料型態保存
3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
# 1. 取出班次一的每一個時間

#個人備註:
#與上面讀取方法一樣
#使用開空LIST，然後把資料等下用PANDAS處理
#指定要第5個，雖然不知道為什麼打5就會自己把5那一列全部用出來...
#=============================================================

import csv

data=[] #開空LIST

with open ('./example.csv',newline="",encoding='utf-8') as file:
    rows= csv.reader(file)#不然其實在這個()裡就可以直接設定了
    for i in rows:
        print(i[5])
        data.append(i)
```

```python
#個人備註:
#丟到一個名為data的空list了
#=============================================================

data
```

```python
import pandas as pd
dt=pd.DataFrame(data)
dt
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存

#個人備註:
#把班次1的時間取出，命名為shift_1的一個空list
#每一個在dt中第5列的班次1時間，全部加進shift_1中
#=============================================================

shift_1=[]

for i in dt[5]:
    shift_1.append(i)
```

```python
shift_1
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存


#個人備註:
#如法炮製，全部開空LIST丟進去
#=============================================================

#shift_1已存在
shift_2=[]
shift_3=[]
shift_4=[]
shift_5=[]

for i in dt[6]:
    shift_2.append(i)
    
for i in dt[7]:
    shift_3.append(i)
    
for i in dt[8]:
    shift_4.append(i)
    
for i in dt[9]:
    shift_5.append(i)
```

```python
#個人備註:
#還不夠結構化
#=============================================================

print(shift_1,shift_2,shift_3,shift_4,shift_5)
```

```python
#個人備註:
#開DICT
#創建一個名為shift的字典
#=============================================================

shift={'shift_':{1:shift_1,
                 2:shift_2,
                 3:shift_3,
                 4:shift_4,
                 5:shift_5,}
       }
```

```python
#個人備註:
#已存入字典，但在查閱時還是不夠好看
#=============================================================

print(shift['shift_'][1])
print(shift['shift_'][2])
print(shift['shift_'][3])
print(shift['shift_'][4])
print(shift['shift_'][5])
```

```python
#個人備註:
#用for的方法把它排出來
#=============================================================
for q,w,e,r,t in zip(shift['shift_'][1],shift['shift_'][2],shift['shift_'][3],shift['shift_'][4],shift['shift_'][5]):
    print(q,w,e,r,t)
```
