```python
#about _thread and threading
#https://www.runoob.com/python3/python3-multithreading.html
import _thread 
import time
import requests
from bs4 import BeautifulSoup
```

```python
startTime = time.time()

r = requests.get('https://www.ettoday.net/news/news-list.htm')
soup = BeautifulSoup(r.text,'lxml')
empty=[]
for i in soup.find('div', class_='part_list_2').find_all('h3'):
    date=i.find('span',class_="date").text
    type_=i.find('em').text
    title=i.find('a').text
    empty.append([date,type_,title])
    
finishTime = time.time()

print(finishTime - startTime)
```

```python
#政治|財經|論壇|國際|大陸，5個不同分類，但網頁格式都一樣，用來測試等下的多線程爬蟲
all_links=['https://www.ettoday.net/news/news-list-2020-04-04-1.htm',
'https://www.ettoday.net/news/news-list-2020-04-04-17.htm',
'https://www.ettoday.net/news/news-list-2020-04-04-13.htm',
'https://www.ettoday.net/news/news-list-2020-04-04-2.htm',
'https://www.ettoday.net/news/news-list-2020-04-04-3.htm']
```

```python
def testing(urls):
    empty=[]
    r = requests.get(urls)
    soup = BeautifulSoup(r.text,'lxml')
    for i in soup.find('div', class_='part_list_2').find_all('h3'):
        date=i.find('span',class_="date").text
        type_=i.find('em').text
        title=i.find('a').text.replace('\u3000','')
        empty.append([date,type_,title])  
    print(empty)
    print('==============================')
```

```python
startTime = time.time() 

for url in all_links:
    testing(url)
    
finishTime = time.time()
print('所花秒數:',round(finishTime - startTime,8))
```

```python
#快超級多???
startTime = time.time() 
for url in all_links:
    _thread.start_new_thread(testing,(url,))    
finishTime = time.time()
print('所花秒數:',round(finishTime - startTime,8))
```
