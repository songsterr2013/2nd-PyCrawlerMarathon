```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'lxml')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
import re
data = {}
new_table = table.find_all('td')

for i in new_table:
    if 'style' in i.attrs:
        item=i.text
        data[item]={}
    if re.compile("\d{4}\/\d{2}").match(i.text):
        date=i.text
        data[item][date]={}
    if re.compile("\d+").match(i.text):
        value=i.text
        data[item][date]=value
data
```
